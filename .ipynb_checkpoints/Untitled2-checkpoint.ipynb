```python
import numpy as np
import glob
import json_tricks as json
import os
```

```python
db_path = '/data/dataset/MuCo-3DHP'
annotations = json.load(os.path.join(db_path, 'MuCo-3DHP.json'))
```

```python
num_joints = 21
root_id = 14

db_img = {}

db_3d = {}
db_2d = {}
db_vis = {}
db_obj = {}
db_img_path = {}

db_bbox = {}
db_cam = {}
```

```python
for annot in annotations['images']:
    img_id = annot['id']
    seq, clip_no, img_name = annot['file_name'].split('/')

    seq_clip_cam = {}
    seq_clip_cam['res_w'] = annot['width']
    seq_clip_cam['res_h'] = annot['height']
    seq_clip_cam['R'] = annot['R']
    seq_clip_cam['t'] = annot['T']
    seq_clip_cam['center'] = annot['c']
    seq_clip_cam['f'] = annot['f']

    db_img[img_id] = [annot['file_name'], seq_clip_cam]


    db_seq_img_path = db_img_path.setdefault(seq, {})
    db_seq_clip_img_path = db_seq_img_path.setdefault(clip_no, [])

    db_seq_clip_img_path.append(os.path.join(db_path, annot['file_name']))
```

```python
prev_img_id = -1
for annot in annotations['annotations']:
    img_id = annot['image_id']

    file_name, seq_clip_cam = db_img[img_id]
    seq, clip_no, img_name = file_name.split('/')

    db_seq_3d = db_3d.setdefault(seq, {})
    db_seq_clip_3d = db_seq_3d.setdefault(clip_no, [])
    db_seq_2d = db_2d.setdefault(seq, {})
    db_seq_clip_2d = db_seq_2d.setdefault(clip_no, [])
    db_seq_vis = db_vis.setdefault(seq, {})
    db_seq_clip_vis = db_seq_vis.setdefault(clip_no, [])

    db_seq_bbox = db_bbox.setdefault(seq, {})
    db_seq_clip_bbox = db_seq_bbox.setdefault(clip_no, [])
    db_seq_cam = db_cam.setdefault(seq, {})
    db_seq_clip_cam = db_seq_cam.setdefault(clip_no, [])

    if img_id == prev_img_id:
        db_seq_clip_2d[-1].append(annot['keypoints_img'])
        db_seq_clip_3d[-1].append(annot['keypoints_cam'])
        db_seq_clip_vis[-1].append(annot['keypoints_vis'])
        db_seq_clip_bbox[-1].append(annot['bbox'])
    else:
        db_seq_clip_2d.append([annot['keypoints_img']])
        db_seq_clip_3d.append([annot['keypoints_cam']])
        db_seq_clip_vis.append([annot['keypoints_vis']])
        db_seq_clip_bbox.append([annot['bbox']])
        db_seq_clip_cam.append(seq_clip_cam)
        prev_img_id = img_id  
```

```python
for seq in db_3d:
    for clip_no in db_3d[seq]:
        num_frames = len(db_3d[seq][clip_no])

        db_seq_obj = db_obj.setdefault(seq, {})
        seq_clip_obj = db_seq_obj.setdefault(clip_no, [])

        seq_clip_kp3d = []
        seq_clip_kp2d = []
        seq_clip_vis = []

        for frame_no in range(num_frames):
            seq_clip_kp3d_temp = np.zeros((max_num_people, num_joints, 3), dtype='float32')
            seq_clip_kp2d_temp = np.zeros((max_num_people, num_joints, 2), dtype='float32')
            seq_clip_vis_temp = np.zeros((max_num_people, num_joints), dtype='bool')
            seq_clip_obj_temp = np.zeros((max_num_people,), dtype='bool')

            img_width = db_cam[seq][clip_no][frame_no]['res_w']
            img_height = db_cam[seq][clip_no][frame_no]['res_h']

            persons = db_3d[seq][clip_no][frame_no]

            root_depths = [person[root_id][2] for person in persons]
            closest_pid = root_depths.index(min(root_depths))

            ## Remain valid person ids
            _pid_list = [closest_pid]
            for p1 in range(len(persons)):
                if p1 == closest_pid:
                    continue
                picked = True
                for p2 in range(len(persons)):
                    if p1 == p2:
                        continue
                    dist = (np.array(persons[p1][root_id]) - np.array(persons[p2][root_id])) ** 2
                    dist_2d = np.sqrt(np.sum(dist[:2]))
                    dist_3d = np.sqrt(np.sum(dist))
                    if dist_2d < 500 or dist_3d < 500:
                        picked = False
                if picked:
                    _pid_list.append(p1)

            pid_list = []
            for pid in _pid_list:
                p_bbox = db_bbox[seq][clip_no][frame_no][pid]

                x,y,w,h = p_bbox
                x1 = np.max((0, x))
                y1 = np.max((0, y))
                x2 = np.min((img_width-1, x1 + np.max((0, w-1))))
                y2 = np.min((img_height-1, y1 + np.max((0, h-1))))

                if w*h > 0 and x2 >= x1 and y2 >= y1:
                    pid_list.append(pid)

            for i, pid in enumerate(pid_list):
                seq_clip_kp3d_temp[i] = persons[pid]
                seq_clip_kp2d_temp[i] = db_2d[seq][clip_no][frame_no][pid]
                seq_clip_vis_temp[i] = np.ones(num_joints,)
                seq_clip_obj_temp[i] = True

            seq_clip_kp3d.append(seq_clip_kp3d_temp)
            seq_clip_kp2d.append(seq_clip_kp2d_temp)
            seq_clip_vis.append(seq_clip_vis_temp)
            seq_clip_obj.append(seq_clip_obj_temp)

        db_3d[seq][clip_no] = seq_clip_kp3d
        db_2d[seq][clip_no] = seq_clip_kp2d
        db_vis[seq][clip_no] = seq_clip_vis
        db_obj[seq][clip_no] = seq_clip_obj
```
